Making Altair plots to save to Jekyll


```python
import pandas as pd
import numpy as np
```

```python

import altair as alt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv')
```

```python
df
```

```python
dfsmall=df.head(100)
```

```python
chart2 = alt.Chart(dfsmall).mark_bar().encode(
    x=alt.X('Year Constructed:Q', bin=True),
    y='count()',
    color='Agency Name:N'   # or whatever your grouping column is
)
chart2
```
